```python
import pandas as pd
import numpy as np
# Carregando o dataset de clientes 
clientes_df = pd.read_csv('datasets/olist_customers_dataset.csv')
clientes_df.drop_duplicates(keep='first') #remove linhas duplicadas e mantêm a primeira ocorrência
clientes_df
```

```python
print('Contagem de valores nulos:')
print(clientes_df.isnull().sum()) # Conta todos os nulos do dataframe
```

```python
display(f'10 Primeiras linhas:')
print(clientes_df.head(10)) #Mostra as 10 primeiras linhas do dataframe
```

```python
print('Formato da tabela(linhas, colunas):')
print(f'Linhas: {clientes_df.shape[0]}\nColunas: {clientes_df.shape[1]}')
```

```python
print(f'Informações gerais:')
print(clientes_df.info())
```

```python
print("Descrição estatística do DataFrame:")#Somente para dados numéricos
print(clientes_df.describe())
```

```python
#Pegando estado mais frequente e contando com Numpy
estados = clientes_df['customer_state'].to_numpy() 
contador = np.unique(estados, return_counts=True)
indice_mais_comum = np.argmax(contador[1])
print(f'O estado com mais pedidos é {contador[0][indice_mais_comum]} com {contador[1][indice_mais_comum]} pedidos.')
print('-*'*30)
#Pegando cidade mais frequente e contando com Pandas
cidades = clientes_df['customer_city'].value_counts() #Número de pedidos por cidade
print(f'A cidade com mais pedidos foi {cidades.idxmax()} com {cidades.max()} pedidos')
print('-*'*30)
print('Número de pedidos por cidade:')
print(cidades)
```

```python
#Número de pedidos por estado
print('Número de pedidos por estado:')
estado = clientes_df['customer_state'].value_counts()
print(estado)
```
